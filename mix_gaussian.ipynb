```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
```

```python
Xmat = np.genfromtxt('clusterdata.csv', delimiter=',')
K = 3
```

```python
plt.scatter(Xmat[:, 0], Xmat[:, 1], s=10)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Scatter Plot of Xmat')
plt.show()
```

```python
from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(n_components=K, covariance_type='full', random_state=0)
gmm.fit(Xmat)
labels = gmm.predict(Xmat)

plt.scatter(Xmat[:, 0], Xmat[:, 1], c=labels, s=10, cmap='viridis')
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('GMM Clustering of Xmat')
plt.show()
```

```python
weights = gmm.weights_
print(weights)
```
